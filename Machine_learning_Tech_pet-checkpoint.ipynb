```python
def count_unique(df3, cols):
    for col in cols:
        print("\n" + "for column" + col)
        print(df3[col].value_counts())
cat_cols= ["FirstName","MiddleName","LastName","AddressLine1",\
          "City", "StateProvinceName","CountryRegionName", "Education","Occupation","MaritalStatus",\
          "Gender"]
count_unique(df3, cat_cols)
```

```python
def plot_bars(df3,cols):
    for col in cols:
        fig= plt.figure(figsize=(6,6))
        ax= fig.gca()
        counts= df3[col].value_counts()
        counts.plot.bar(ax= ax, color= "red")
        ax.set_title("Number of Individuals by" + col)
        ax.set_xlabel(col)
        ax.set_ylabel("Number of individuals")
        plt.show()
plot_cols = ["MaritalStatus", "Gender", "Occupation", "Education", "CountryRegionName"]
plot_bars(df3, plot_cols)
```

```python
def plot_box(df3, cols, col_y= "AveMonthSpend"):
    for col in cols:
        sns.set_style("whitegrid")
        sns.boxplot(col, col_y, data= df3)
        plt.xlabel(col)
        plt.ylabel(col_y)
        plt.show()
cat_cols= ["FirstName", "Education", "Occupation", "CountryRegionName","MaritalStatus", "Gender"]
plot_box(df3, cat_cols)
```

```python
def plot_histogram(df3,cols, bins = 8):
    for col in cols:
        fig= plt.figure(figsize=(6,6))
        ax= fig.gca()
        df3[col].plot.hist(ax=ax, bins= bins, color = "orange")
        ax.set_title("Histogram of" + col)
        ax.set_xlabel(col)
        ax.set_ylabel("number of individuals")
        plt.show()
num_cols= ["YearlyIncome", "TotalChildren", "NumberCarsOwned", "NumberChildrenAtHome"]
plot_histogram(df3, num_cols)
```

```python
def plot_den_hist(df3,cols, bins = 10, hist =True):
    for col in cols:
        sns.set_style("whitegrid")
        sns.distplot(df3[col], bins = bins, rug = True, hist= hist)
        plt.title("Histogram of " + col)
        plt.xlabel(col)
        plt.ylabel("number of individuals")
        plt.show()

plot_den_hist(df3, num_cols)
```

```python
def plot_scatter(df3, cols, col_y = "AveMonthSpend"):
    for col in cols:
        fig = plt.figure(figsize=(7,8))
        ax = fig.gca()
        df3.plot.scatter(x = col, y= col_y, ax = ax)
        ax.set_title("scatter plot of" + col_y + "vs" + col)
        ax.set_xlabel(col)
        ax.set_ylabel(col_y)
        plt.show()
plot_scatter(df3, num_cols)
```

```python
def plot_density_2d(df3, cols, col_y = "AveMonthSpend", kind ="hex"):
    for col in cols:
        sns.set_style("whitegrid")
        sns.jointplot(col, col_y, data= df3, kind= kind)
        plt.xlabel(col)
        plt.ylabel(col_y)
        plt.show()
        
plot_density_2d(df3, num_cols)
```

```python
def plot_box(df3, cols, col_y = 'AveMonthSpend'):
    for col in cols:
        sns.set_style("whitegrid" ) 
        sns.boxplot(col, col_y, data=df3)
        plt.xlabel(col) # Set text for the x axis
        plt.ylabel(col_y)# Set text for y axis
        plt.show()
        
cat_cols = ["City", "FirstName", "Education", "Occupation", "CountryRegionName","MaritalStatus", "Gender"]
plot_box(df3, cat_cols) 
```
